# Online News Popularity Dataset

<h3>I would like to look at data from an online news site to determine how different aspects of each news article affect its popularity, expressed by number of shares.</h3>
<h4>https://archive.ics.uci.edu/ml/datasets/Online+News+Popularity </h4>

This set of data provided by UCI's Machine Learning Repository describes a number of news articles posted on Mashable.com before January 8, 2015. With 39644 data entries and 61 descirptive attributes, there is plenty to compare. The attribute types vary, relating to content, length, keywords, when the article was posted, and sentiment. I am especially interested in how sentiment in the text compares to the number of shares. Ultimately we may find what helps make a news posting more shareable, and what is important or not important to focus on when writing and putting together a new article.

### This was my original goal, to guide what Mashable.com should and should not spend time and energy on in creating their articles. I wanted to be able to deliver a plan to help the company gain a greater amount of shares. My first pass at the data, I thought I was giving great advice, but in the reformatting of my presentation, incuding the use of one method, df.corr(), the data was put into a different perspective. 

#### Let's see what happened.

```python
import numpy as np
import pandas as pd
from scipy.stats import iqr
from ipywidgets import interact, IntSlider
from IPython.display import display
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
import seaborn as sns
%matplotlib inline
from pylab import hist, show, xticks
from IPython.lib.display import YouTubeVideo
```

 We clean up the data a little first. Import the .csv file and strip column names of whitespace.

```python
original = pd.read_csv('online_news_popularity.csv')
news = original.rename(columns=lambda x: x.strip())
```

```python
pd.set_option('display.max_colwidth', -1)
#This code shows the full column widths so the urls are not truncated.
```

### With a correlation table, we can see which attributes are clearly affecting each other. This table is color coded for easier reading because there are so many attributes (60!). Any warm colored to deep red cells indicate a strong correlation.

```python
news.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'))
```

#####  Cells with "1.0" we can see indicate the attribute correlates 100%, which usually means the attributes was compared with itself (see the diagonal red line of cells going through the table?), so we ignore those. However, n_unique_tokens,	n_non_stop_words, and	n_non_stop_unique_tokens all match at 100% with each other....

## Most importantly, we can see the last column for 'shares' is all a very deep blue. Cue record scratch. The numbers listed are not very close to 1 at all, showing there is little correlation with all the other attributes and shares. So if we thought we could glean some guidance as to what aspects of an article affect the shares, we can't because that's not something the data is showing us!

### But we don't have to throw the baby out with the bathwater. There is some correlation here and there, let's check them out.

### We can see the attributes indicating the genre of the article, Lifestyle, Entertainment, Business, Social Media, Tech, and World, are generally correlated with the LDA attributes. Other than that, a lot of the correltations are between attributes that are definitely already related to each other, and perhaps even formed from each other, based on their names and desrciptions provided onthe web link.

### LDA topics are noted as 0 through 4, but we don't know to what they are referring. What is LDA anyway?

```python
lda_and_genre = news[['LDA_00',
                  'LDA_01',
                  'LDA_02',
                  'LDA_03',
                  'LDA_04',
                  'data_channel_is_lifestyle',
                  'data_channel_is_entertainment',
                  'data_channel_is_bus',
                  'data_channel_is_socmed',
                  'data_channel_is_tech',
                  'data_channel_is_world']].copy()
lda_and_genre.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'))
```

```python
YouTubeVideo('DWJYZq_fQ2A')
```

In general, Latent Dirichlet Allocation determines the occurrance of topics over mutiple texts and compares the occurances of topics between them.

Another exmple of correlations that are not helpful:
We see here, of course, that global rate of negatives and positives correlate with rate of positive or negative words.
Definitions include:
- global_sentiment_polarity: Text sentiment polarity 
- global_rate_positive_words: Rate of positive words in the content 
- global_rate_negative_words: Rate of negative words in the content 
- rate_positive_words: Rate of positive words among non-neutral tokens 
- rate_negative_words: Rate of negative words among non-neutral tokens

```python
sentiment = news[['global_subjectivity',
                  'global_sentiment_polarity',
                  'global_rate_positive_words',
                  'global_rate_negative_words',
                  'rate_positive_words',
                  'rate_negative_words']].copy()
sentiment.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'))
```

## There is not much to be gained from the correlation tables at first glance; they are unhelpful in determining how to affect more shares, and do not give a clear direction with any comparisons between two given attributes, it is unclear how to proceed.
